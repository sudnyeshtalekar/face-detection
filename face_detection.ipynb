```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('IMG_3941.JPG')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
detector = cv2.CascadeClassifier('ml-vit-vellore-2019-master/Datasets/FaceCascade/templatedata.xml')
```

```python
faces = detector.detectMultiScale(img,1.1)
```

```python
for f in faces:
    x,y,w,h = f
    green = (0,255,0)
    cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    
plt.imshow(img)
plt.show()
```

```python
camera = cv2.VideoCapture(0)

while True:
    b,img = camera.read()
    if b==False:
        continue
    # Find out faces--------------------------
    faces = detector.detectMultiScale(img,1.3)
    
    for f in faces:
        x,y,w,h = f
        green = (0,255,0)
        cv2.rectangle(img,(x,y),(x+w,y+h),green,5)
    #------------------------------------------
    cv2.imshow("Image ",img)
    key = cv2.waitKey(1)&0xFF
    if key==ord('q'):
        break
    
camera.release()   
cv2.destroyAllWindows()
    
```
