````python
x = np.linspace(0,11,100)
y = np.sin(x)
plt.plot(x,y,label="sin(X)",color='red')
plt.plot(x,np.cos(x),label="cos(X)")
plt.title("Sin vs Cos")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.show()
````
